## 34.Requestsの解説

### Responseオブジェクトの属性
- status_code: ステータスコード
- headers: レスポンスヘッダの内容
- contnet: レスポンスのバイナリデータ
- text: レスポンスの内容
- encoding: エンコーディング
- cookies: クッキーの内容

### ステータスコード
- 100番台: リクエストは受け入れられ、処理が継続
- 200番台: リクエストに成功
- 300番台: リダイレクトなどリクエストの完了には追加的な処理が必要
- 400番台: Client Error
- 500番台: Server Error

## 35. Requestsの使い方(responseオブジェクト)

```python
# requestsをインポート
import requests
```

```python
#　urlからデータ取得
response = requests.get('https://www.yahoo.co.jp/')
```

```python
# ステータス状況を確認、200番台だと接続成功
response.status_code
```

```python
# htmlの確認
response.text
```

```python
# サーバとのやり取りをしたバイナリデータ
response.content
```

```python
# 使用しているencodeingを確認
response.encoding
```

```python
# headerの確認,辞書型に返ってくる
response.headers
```

```python
for key, val in response.headers.items():
    print(key, ' ',val)    
```

```python
# クッキーを確認
response.cookies
```

## 36.Requestsの使い方(getメソッドの引数)

```python
user_agent ='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'
```

```python
header = {'user-agent': user_agent}
```

```python
url = 'https://www.yahoo.co.jp/'
```

```python
# headerを指定
response = requests.get(url, headers = header)
```

```python
response.status_code
```

```python
# timeoutを指定
response = requests.get(url, timeout=3)    
```

```python
response.text[:500]
```

```python
# サーバに情報を渡すためにクエリパラメータを使う
param = {'q' : 'python'}
```

```python
response = requests.get('https://www.google.com/search', params = param)
```

```python
response.text[:1000]
```

## 45.読売新聞オンラインからHTMLデータの取得

```python
# モジュールのインポート
import requests
from bs4 import BeautifulSoup
```

```python
# requestsでhtmlを取得
URL ='https://www.yomiuri.co.jp/'
res = requests.get(URL)
```

```python
# 接続確認
res.status_code
```

```python
# htmlの中身確認
res.text
```

## 46. selectメソッドによるニュースタイトル・URLの取得

```python
# BeatuifulSoupに必要なデータを渡す
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# selectメソッドで取得
elems = soup.select('body > div.uni-home-2020.home-2020-wrap > div.home-2020-primebox > div.home-2020-primebox-article > div.home-2020-prime-topnews.active > h3 > a')
```

```python
# a要素のタイトルとURLを取得
elems[0]
```

```python
# 中身を確認するにはcontents(contensもリストになるのでスライスする)
elems[0].contents[0]
```

```python
# 要素が含まれる属性を抽出する場合
elems[0].attrs['href']
```

## 47. 複数のニュースタイトル・URLの取得

```python
# 複数のCSSセレクターを確認し、同じ部分を記載する
elems = soup.select('body > div.uni-home-2020.home-2020-wrap > div.home-2020-primebox > div.home-2020-primebox-article > div.home-2020-prime-headline.active > ul')
```

```python
# elemsはリストなのでスライスで中身を表示
elems[0]
```

```python
# データを整形
print(elems[0].prettify())
```

```python
# BeautifulSoupのタグオブジェクトを確認
# タグオブジェクトは.メソッドで中身のtagを選択可能
print(type(elems[0]))
```

```python
# h3要素を取得
elems[0].h3
```

```python
# textを取得
elems[0].h3.a.string
```

```python
# 属性を取得
elems[0].h3.a['href']
```

```python
# 次の表題を取得する
elems[0].li.next_sibling.next_sibling
```

```python
# ニュースのタイトルを取得
elems[0].li.next_sibling.next_sibling.h3.a.string
```

```python
# URLを取得
elems[0].li.next_sibling.next_sibling.h3.a['href']
```

```python
# next_siblingsで複数の要素を取得できる

# htmlに広告文があったので除外する

for sibling in elems[0].li.next_siblings:
    # h3タグの内容を変数tag_h3に格納（改行\nの場合は、空白）
    tag_h3 = sibling.h3 if sibling != "\n" else ""
    # h3タグが含まれ(not None)、かつ、空白で無い場合
    if tag_h3 is not None and tag_h3 != "":
        print(tag_h3.a.string)
        print(tag_h3.a["href"])
```

## 48. find_allメソッドによるニューススタイル・URLの取得

```python
elems_news = elems[0].find_all('h3')
elems_news
```

```python
# for文を使って取得
for elem in elems_news:
    print(elem.a.string)
    print(elem.a['href'], end='\n\n')
```
